# LETS ANALIZE EACH BRANCHE AND ITS CUSTOMERS

```sql
SELECT * FROM dbo.supermarket_sale WHERE Branch = 'A';
```

```sql
SELECT Customer_type, Gender, COUNT(Customer_type) AS 'Total_number' FROM dbo.supermarket_sale WHERE Branch = 'A' GROUP BY Customer_type, Gender ORDER BY Total_number;
```

```sql
SELECT Product_line, Gender,
COUNT(Gender) AS 'Total_number' FROM dbo.supermarket_sale 
WHERE Branch = 'A' GROUP BY Product_line, Gender ORDER BY Total_number DESC;
```

```sql
SELECT Product_line, Gender, 
ROUND(AVG(Unit_price), .1) AS 'AVG_price' FROM dbo.supermarket_sale 
WHERE Branch = 'A' GROUP BY Product_line, Gender ORDER BY AVG_price DESC
```

```sql
/* POPULAR PAYMENT METHOD FOR EACH OF THE BRANCHCES */
/* PIVOT TABLE SHOWING THE DISTRIBUTION */
SELECT * FROM
(
    SELECT Branch, Payment FROM dbo.supermarket_sale
) AS SourceTable
PIVOT
(
    COUNT(Payment)
    FOR Payment IN (
        [Ewallet], [Credit card], [Cash]
    )
) AS Pivot_table
```

```sql
/*CONSTRUCT A DYNAMIC SQL PIVOT TABLE 
DECLARE
    @columns NVARCHAR(MAX) = '', 
    @sql       NVARCHAR(MAX) = '';
SELECT 
    @columns+= QUOTENAME(Payment) + ','
FROM dbo.supermarket_sale
SET @columns = LEFT(@columns, LEN(@columns) - 1);

-- construct dynamic SQL
SET @sql ='
SELECT * FROM   
(
    SELECT 
        Branch, Payment 
        FROM dbo.supermarket_sale
) AS SOURCE_TABLE
PIVOT(
    COUNT(Payment) 
    FOR Payment IN ('+ @columns +')
) AS pivot_table;';
-- execute the dynamic SQL
EXECUTE sp_executesql @sql;
*/
```

```sql
/* GETTING THE TOTAL CUTOMERS THAT VISITED STORE A */
WITH branch AS (
    SELECT Branch, Date AS 'Date_of_month' FROM 
    dbo.supermarket_sale
    WHERE Branch ='A'
), branch_visitor AS(
    /* FINDING THE INTERVALS BETWEEN BRANCH A AND ITS TOTAL USERS*/
    SELECT CASE
        WHEN (Date_of_month BETWEEN '2019-01-01' AND '2019-01-31')
         THEN 1
        ELSE 0
        END AS 'First_month',
    CASE 
        WHEN(Date_of_month BETWEEN '2019-02-01' AND '2019-02-28') THEN 1
        ELSE 0
        END AS 'Second_month',
    CASE 
        WHEN(Date_of_month BETWEEN '2019-03-01' AND '2019-03-31') THEN 1
        ELSE 0
        END AS 'Third_month'
    FROM branch
)
/* GETTING THE TOTAL CUTOMERS THAT VISITED STORE A */
SELECT SUM(First_month) AS 'January_CUS', SUM(Second_month) AS 'Febuary_cus',
SUM(Third_month) AS 'March_cus' 
FROM branch_visitor;
```
